## Customer Loyalty Prediction Model

```python
import numpy as np
import matplotlib.pyplot as plt

#   Matrix A
#      A   B
#  A [[], []]
#  B [[], []]

store_a = float(input("The probability of the customers shopping at Store A again the following week (%): ")) / 100
store_b = float(input("The probability of the customers shopping at Store B again the following week (%): ")) / 100

a = np.matrix([[store_a, 1-store_b],
     [1 - store_a, store_b]])

customers_a = int(input("How many customers does Store A have: "))
customers_b = int(input("How many customers does Store B have: "))

s0 = np.matrix([[customers_a],  # Column Matrix
                [customers_b]])

weeks = int(input("How many weeks do you want to predict?: "))

for i in range(weeks):
    s0 = a * s0
    
    
store_a_prediction = int(s0.item(0))
store_b_prediction = int(s0.item(1))

x = ["Store A", "Store B"]
y = [store_a_prediction, store_b_prediction]
colors = ["blue", "green"]

total = y[0] + y[1]

store_a_percent = (y[0] / total) * 100
store_b_percent = (y[1] / total) * 100

def add_labels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')
        
add_labels(x,y)
plt.bar(x, height = y, color = colors)
plt.xticks(x)
plt.ylabel("Customers")
plt.xlabel("Stores")
plt.title("After " + str(weeks) + " weeks")

plt.show()
```
