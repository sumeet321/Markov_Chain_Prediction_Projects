#### Commuters traveling from Pomona to Los Angeles who take I-10 are 60% likely to take it the next day, 30% likely to take CA-60, and 10% likely to take I-210. Those who take CA-60 one day, are 70% likely to take it the next day, 10% likely to take I-10, and 20% likely to take I-210. Drivers of I-210 on one day will take it the next day 80% of the time, and take I-10 the other 20% of the time.

### How many people are commuting on each freeway depending on the amount of days

```python
import numpy as np
import matplotlib.pyplot as plt

#          A Matrix
#       10    60   210 
# 10  [[.6], [.1], [.2]]
# 60  [[.3], [.7], [0]]
# 210 [[.1], [.2], [.8]]

a = np.matrix([
    [.6, .1, .2], 
    [.3, .7, 0], 
    [.1, .2, .8]])

freeway_input_1 = int(input("Amount of people commuting on the I-10 Freeway: "))
freeway_input_2 = int(input("Amount of people commuting on the CA-60 Freeway: "))
freeway_input_3 = int(input("Amount of people commuting on the I-210 Freeway: "))

s0 = np.matrix([[freeway_input_1],
                [freeway_input_2],
                [freeway_input_3]])
    
days = int(input("How many days do you want to predict?: "))
for i in range(days):
    s0 = a * s0
        

ten_prediction = int(s0.item(0))
sixty_prediction = int(s0.item(1))
two_ten_prediction = int(s0.item(2))

total_commuters = ten_prediction + sixty_prediction + two_ten_prediction

ten_percent = (ten_prediction / total_commuters) * 100
sixty_percent = (sixty_prediction / total_commuters) * 100
two_ten_percent = (two_ten_prediction / total_commuters) * 100


x = ["I-10", "CA-60", "I-210"]
y1 = [ten_prediction, sixty_prediction, two_ten_prediction]
y2 = [ten_percent, sixty_percent, two_ten_percent]

#def add_labels(x,y):
#    for i in range(len(x)):
#        ax[i].text(i, y[i], y[i], ha = 'center')
        
        
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

#add_labels(x, y1)
#add_labels(x, y2)

ax1.bar(x, y1, color = "Green")
ax1.bar_label(ax1.containers[0]) # Amount of Commuters

ax2.bar(x, y2, color = "Green")
#ax2.bar_label(ax2.containers[1]) # Percentages of Commuters
plt.xticks(x)
ax1.set_ylabel("Commuters", color = "Green")
ax2.set_ylabel("Percentages of Commuters", color = "Blue")
plt.xlabel("Freeways")
ax1.set_title("After " + str(days) + " days")

plt.show()
```
