#### Commuters traveling from Pomona to Los Angeles who take I-10 are 60% likely to take it the next day, 30% like to take CA-60, and 10% likely to take I-210. Those who take CA-60 one day, are 70% likely to take it the next day, 10% likely to take I-10, and 20% likely to take I-210. Drivers of I-210 on one day will take it the next day 80% of the time, and take I-10 the other 20% of the time.

### What is the likelyhood of getting on each freeway depending on the amount of days

```python
import numpy as np
import matplotlib.pyplot as plt

#          A Matrix
#       10    60   210 
# 10  [[.6], [.1], [.2]]
# 60  [[.3], [.7], [0]]
# 210 [[.1], [.2], [.8]]

a = np.matrix([
    [.6, .1, .2], 
    [.3, .7, 0], 
    [.1, .2, .8]])

choice = input("What freeway did the driver take today? (10), (60), (210): ")
if choice == "10":
    s0 = np.matrix([[1], [0], [0]]) # Column Matrix
elif choice == "60":
    s0 = np.matrix([[0], [1], [0]]) # Column Matrix
elif choice == "210":
    s0 = np.matrix([[0], [0], [1]]) # Column Matrix
else:
    print("Wrong input. Please try again!")
    quit()
    
days = int(input("How many days do you want to predict?: "))
for i in range(days):
    s0 = a * s0
    
ten_prediction = s0.item(0) * 100
sixty_prediction = s0.item(1) * 100
two_ten_prediction = s0.item(2) * 100

predictions = [ten_prediction, sixty_prediction, two_ten_prediction]

freeways = ["I-10", "CA-60", "I-210"]

plt.bar(freeways, predictions)
plt.xticks(freeways)
plt.ylabel("Percentages (%)")
plt.xlabel("Freeways")
plt.title("After " + str(days) + " days")

plt.show()
```
